```python
import os

# 1. Predefined dictionary of stock prices (As specified in instructions)
STOCK_PRICES = {
    "AAPL": 180,
    "TSLA": 250,
    "NVDA": 120,
    "MSFT": 420,
    "AMZN": 180
}

def add_stock(user_portfolio):
    """Prompts user to input stock name and quantity, adding it to their portfolio."""
    print("\n--- Add/Update Stock Holdings ---")
    ticker = input("Enter stock symbol (e.g., AAPL, TSLA): ").upper().strip()
    
    if ticker in STOCK_PRICES:
        try:
            quantity = int(input(f"Enter number of shares for {ticker}: "))
            if quantity < 0:
                print("Quantity cannot be negative.")
                return
            
            # Update portfolio dictionary with user input
            if ticker in user_portfolio:
                user_portfolio[ticker] += quantity
            else:
                user_portfolio[ticker] = quantity
            print(f"Added {quantity} shares of {ticker} successfully!")
            
        except ValueError:
            print("Invalid input. Please enter a whole number for quantity.")
    else:
        print("Stock symbol not found in our predefined price list.")

def display_portfolio(user_portfolio):
    """Calculates and displays the breakdown and total investment value."""
    print("\n================ CURRENT PORTFOLIO ================")
    if not user_portfolio:
        print("Your portfolio is currently empty.")
        print("==================================================")
        return
    
    total_investment = 0
    print(f"{'Stock':<10}{'Shares':<10}{'Price/Share':<15}{'Total Value':<12}")
    print("-" * 50)
    
    for ticker, quantity in user_portfolio.items():
        price = STOCK_PRICES[ticker]
        stock_value = quantity * price
        total_investment += stock_value
        print(f"{ticker:<10}{quantity:<10}${price:<14}${stock_value:<12}")
        
    print("-" * 50)
    print(f"Total Investment Value: ${total_investment}")
    print("==================================================")

def main():
    # User's empty portfolio dictionary to store tracking state
    user_portfolio = {}
    
    while True:
        print("\n📈 Stock Tracker Menu:")
        print("1. Add / Update Stock")
        print("2. View Total Investment Value")
        print("3. Exit")
        
        choice = input("Select an option (1-3): ").strip()
        
        if choice == "1":
            add_stock(user_portfolio)
        elif choice == "2":
            display_portfolio(user_portfolio)
        elif choice == "3":
            print("\nExiting Stock Tracker. Happy investing!")
            break
        else:
            print("Invalid selection. Please choose option 1, 2, or 3.")

if __name__ == "__main__":
    main()
```
