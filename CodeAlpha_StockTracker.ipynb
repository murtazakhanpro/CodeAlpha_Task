```python
# 1. Predefined dictionary of stock prices
stock_market = {
    "AAPL": 180,
    "TSLA": 250,
    "NVDA": 120,
    "MSFT": 420,
    "AMZN": 180
}

# 2. Empty dictionary to track user's holdings
user_portfolio = {}

# 3. The Main Program Loop
while True:
    print("\n📈 Stock Tracker Menu:")
    print("1. Add / Update Stock")
    print("2. View Total Investment Value")
    print("3. Exit")
    
    choice = input("Select an option (1-3): ").strip()
    
    if choice == "1":
        print("\n--- Add/Update Stock Holdings ---")
        ticker = input("Enter stock symbol (e.g., AAPL, TSLA): ").upper().strip()
        
        # Check if the stock symbol exists in our market dictionary
        if ticker in stock_market:
            quantity_input = input(f"Enter number of shares for {ticker}: ")
            
            # Simple validation check to ensure user typed a valid digit
            if quantity_input.isdigit():
                quantity = int(quantity_input)
                
                # Add or update the dictionary tracking state
                if ticker in user_portfolio:
                    user_portfolio[ticker] += quantity
                else:
                    user_portfolio[ticker] = quantity
                print(f"Added {quantity} shares of {ticker} successfully!")
            else:
                print("Invalid input! Please enter a positive whole number.")
        else:
            print("Stock symbol not found in our predefined price list.")
            
    elif choice == "2":
        print("\n================ CURRENT PORTFOLIO ================")
        
        # Check if dictionary is empty
        if len(user_portfolio) == 0:
            print("Your portfolio is currently empty.")
        else:
            total_investment = 0
            print("Stock     Shares    Price/Share    Total Value")
            print("-" * 46)
            
            # Loop through tracking dictionary to calculate arithmetic values
            for ticker in user_portfolio:
                quantity = user_portfolio[ticker]
                price = stock_market[ticker]
                stock_value = quantity * price
                total_investment += stock_value
                
                print(f"{ticker:<10}{quantity:<10}${price:<14}${stock_value}")
                
            print("-" * 46)
            print(f"Total Investment Value: ${total_investment}")
        print("==================================================")
        
    elif choice == "3":
        print("\nExiting Stock Tracker. Happy investing!")
        break  # Safely breaks out of the while loop to terminate program
        
    else:
        print("Invalid selection. Please choose option 1, 2, or 3.")
```
