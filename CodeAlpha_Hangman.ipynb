```python
import random

class Hangman:
    def __init__(self):
        # 1. Predefined words (Simplified scope: no file/API needed)
        words = ["python", "coding", "intern", "alpha", "script"]
        self.secret_word = random.choice(words)
        
        # 2. Tracking game state
        self.lives = 6  # Limit incorrect guesses to 6
        self.guessed_letters = set()  # Stores unique correct guesses

    def display_word(self):
        """Displays the secret word with underscores for unguessed letters."""
        display = ""
        for letter in self.secret_word:
            if letter in self.guessed_letters:
                display += letter + " "
            else:
                display += "_ "
        print("\nWord: " + display.strip())
        print(f"Incorrect guesses remaining: {self.lives}")

    def play(self):
        """Main game loop managing turns, inputs, and win/loss states."""
        print("Welcome to Hangman!")
        
        # Loop runs while player has lives and hasn't won yet
        while self.lives > 0 and set(self.secret_word) != self.guessed_letters:
            self.display_word()
            
            # Get user input and normalize to lowercase
            guess = input("Guess a letter: ").lower().strip()
            
            # Input validation: Ensure it's a single letter
            if len(guess) != 1 or not guess.isalpha():
                print("Please enter a single valid alphabet letter.")
                continue
                
            # Process the guess
            if guess in self.secret_word:
                if guess in self.guessed_letters:
                    print(f"You already guessed '{guess}' correctly!")
                else:
                    print(f"Good job! '{guess}' is in the word.")
                    self.guessed_letters.add(guess)
            else:
                print(f"Sorry, '{guess}' is not in the word.")
                self.lives -= 1  # TODO: Deduct a life here! (Completed)
                
        # Game Over checking
        if set(self.secret_word) == self.guessed_letters:
            print(f"\n🎉 Congratulations! You guessed the word: {self.secret_word}!")
        else:
            # TODO: Add the losing message (Completed)
            print("\n❌ Game Over!")
            print(f"You ran out of lives. The correct word was: '{self.secret_word}'")

# To run the game:
if __name__ == "__main__":
    game = Hangman()
    game.play()
```
