```python
print("==================================================")
print("🤖 Welcome! I am your simple rule-based AI chatbot.")
print("Type 'bye' at any time to end our conversation.")
print("==================================================")

# The conversation loop keeps the chatbot awake and listening
while True:
    # Get text input from the user, remove blank spaces, and make it lowercase
    user_message = input("\nYou: ").strip().lower()
    
    # Check the user's message against predefined rules using if-elif-else
    if user_message == "hello" or user_message == "hi" or user_message == "hey":
        print("Chatbot: Hi there! How can I help you today?")
        
    elif user_message == "how are you":
        print("Chatbot: I'm doing great, thank you for asking! How are you?")
        
    elif user_message == "what is your name":
        print("Chatbot: I am a simple Python chatbot built for Task 4!")
        
    elif user_message == "bye" or user_message == "goodbye":
        print("Chatbot: Goodbye! Have a wonderful day!")
        break  # Exits the while loop and closes the program
        
    else:
        # Fallback response if the chatbot doesn't recognize the command
        print("Chatbot: I'm sorry, I don't understand that command yet. Try saying 'hello', 'how are you', or 'bye'.")
```
