```python
import os
import shutil

print("==================================================")
print("📂 STEP 1: SETTING UP DIRECTORIES")
print("==================================================")

# Define the source folder (where your messy files currently are)
source_folder = "MyDocuments"

# Define the destination folder (where you want to organize the JPGs)
destination_folder = "OrganizedPhotos"

# Automatically create the destination folder if it doesn't exist yet
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)
    print(f"Created new folder: '{destination_folder}'")

# Create a dummy source folder with sample files if you don't have one, just to test!
if not os.path.exists(source_folder):
    os.makedirs(source_folder)
    # Creating empty files inside to simulate a real folder setup
    with open(f"{source_folder}/photo1.jpg", "w") as f: f.write("")
    with open(f"{source_folder}/photo2.jpg", "w") as f: f.write("")
    with open(f"{source_folder}/notes.txt", "w") as f: f.write("")
    print(f"Created temporary '{source_folder}' with sample files for testing.")

print("==================================================")
print("🚀 STEP 2: RUNNING THE AUTOMATION LOOP")
print("==================================================")

# Get a list of every single file name inside the source folder
files_list = os.listdir(source_folder)

moved_count = 0

# Loop through every file name one by one
for file_name in files_list:
    
    # Check if the file ends with the .jpg extension
    if file_name.lower().endswith(".jpg"):
        
        # Construct the complete file paths
        old_path = os.path.join(source_folder, file_name)
        new_path = os.path.join(destination_folder, file_name)
        
        # Move the file from old path to new path
        shutil.move(old_path, new_path)
        
        print(f"Moved: {file_name} ➔ {destination_folder}/")
        moved_count += 1

print("--------------------------------------------------")
print(f"Automation complete! Total files successfully moved: {moved_count}")
print("==================================================")
```
